```python
import io

import pandas as pd
from matplotlib import pyplot as plt
movie = pd.read_csv('movies.csv')
writer_scores = movie.groupby('writer')['score'].mean().reset_index()
writer_scores_limited = writer_scores.head(100)

plt.scatter(range(len(writer_scores_limited)), writer_scores_limited['score'],
            marker="x", color='green', alpha=0.7, s=100)
plt.xticks(range(len(writer_scores_limited)), writer_scores_limited['writer'],
           rotation=45, ha='right')
plt.xlabel('Writer')
plt.ylabel('Mean Score')
plt.title('Mean Score by Writer (Top 30)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import glob
import os

csv_directory = "C:/Users/dai/PycharmProjects/Day15/tobemerged"

all_filenames = glob.glob(os.path.join(csv_directory, "*.csv"))

dfs = []

for filename in all_filenames:
    df = pd.read_csv(filename)
    dfs.append(df)

combined_df = pd.concat(dfs, ignore_index=True)

combined_df.to_csv("merged_output.csv", index=False)

print("CSV files merged successfully!")
```

```python
import pandas as pd
df = pd.read_csv("ecommerce_app_detailed.log", delimiter = "\t", names= ["log_line"], header=None)
df.head()
```

```python
df_split = df['log_line'].str.split(' ', n=2, expand=True)
# df['Timestamp_Raw'] = df_split[0]
# df['remaining'] = df_split[1]
df_split.head()
```

```python
df_split[0] = df_split[0] + " " + df_split[1]
df_split.drop(1, axis = 1, inplace=True)
df['Timestamp_Raw'] = df_split[0]
df['remaining'] = df_split[2]
df_split.head()
```
